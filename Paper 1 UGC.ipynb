```python
from sklearnex import patch_sklearn
patch_sklearn()
 
import os
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.svm import SVC 

loc = "C:/Users/taman/Desktop/Paper 1 UGC/Dataset"
res = "C:/Users/taman/Desktop/Paper 1 UGC/Results"

mlp = MLPClassifier(max_iter =2000,tol = 0.0001)
dt = DecisionTreeClassifier()
gnb = GaussianNB()
mnb = MultinomialNB()
knn = KNeighborsClassifier()
gpc = GaussianProcessClassifier()
sgd = SGDClassifier()

svm = SVC()
sm = SMOTE()
ss = MinMaxScaler()
ss1 =StandardScaler()

clfs = [mlp, dt,gnb,mnb,knn,gpc,sgd,svm]

for file in os.listdir(loc):
    df = pd.read_csv(os.path.join(loc,file))
    X = df.iloc[:,:-1]
    y = df.iloc[:,-1]
    
    p = pd.DataFrame()
    clf1=[]
    for clf in clfs:
        acc1 = []
        
        
        pl = Pipeline ([('smote',SMOTE()), 
                        ('scaler',StandardScaler()),
                        ('classifier',clf)])
        acc = cross_val_score(pl,X,y,cv =10, scoring = 'accuracy').mean()
        roc = cross_val_score(pl,X,y,cv =10, scoring = 'roc_auc').mean()
        rec = cross_val_score(pl,X,y,cv =10, scoring = 'recall').mean()
        pre = cross_val_score(pl,X,y,cv =10, scoring = 'precision').mean()
        fm = cross_val_score(pl,X,y,cv =10, scoring = 'f1').mean()
        
        clf1.append(str(clf).split("(")[0])
        acc1.append(acc)
        acc1.append(roc)
        acc1.append(rec)
        acc1.append(pre)
        acc1.append(fm)
        
        p = pd.concat([p,pd.Series(acc1)], ignore_index = True, axis =1)
    p.index = ['Accuracy','ROC','Recall','Precision','F1Score']
    p.columns = clf1
    p.to_csv(os.path.join(res,file))
```
